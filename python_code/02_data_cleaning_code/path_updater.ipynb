```python
import os
import json

# Define the root directory containing your Jupyter notebooks.
root_dir = "/Users/vk365/Dropbox/Interconnections_data/data/ic_studies/raw/code/phase_1_cost_data/"

# Define the replacements without any extra quotes in the paths.
replacements = {
    r"/Users/vk365/Dropbox/Interconnections_data/data/pdf_scraper/output/":
        r"/Users/vk365/Dropbox/Interconnections_data/data/ic_studies/raw/04_intermediate_scraped_data/phase_1_cost_data/",
    r"/Users/vk365/Dropbox/Interconnections_data/data/pdf_scraper/data":
        r"/Users/vk365/Dropbox/Interconnections_data/data/ic_studies/raw/03_data"
}

# Recursively traverse the directory.
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith(".ipynb"):
            file_path = os.path.join(dirpath, filename)
            print(f"Processing: {file_path}")

            # Load the notebook as JSON.
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    notebook = json.load(f)
            except json.JSONDecodeError as e:
                print(f"Error loading {file_path}: {e}")
                continue

            modified = False

            # Process each cell in the notebook.
            for cell in notebook.get("cells", []):
                if "source" in cell:
                    # The source may be a list or a single string.
                    if isinstance(cell["source"], list):
                        cell_source = "".join(cell["source"])
                    else:
                        cell_source = cell["source"]

                    original_source = cell_source

                    # Perform replacements.
                    for old, new in replacements.items():
                        cell_source = cell_source.replace(old, new)

                    # If any change was made, update the cell's source.
                    if cell_source != original_source:
                        modified = True
                        if isinstance(cell["source"], list):
                            cell["source"] = [line + "\n" for line in cell_source.splitlines()]
                        else:
                            cell["source"] = cell_source

            # Write back the updated notebook if changes were made.
            if modified:
                with open(file_path, "w", encoding="utf-8") as f:
                    json.dump(notebook, f, indent=1)
                print(f"Updated {file_path}")
            else:
                print(f"No changes in {file_path}")
```
